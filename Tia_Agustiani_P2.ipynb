# Definisi Deep Learning

Deep Learning adalah bagian dari Machine Learning yang mengkhususkan diri dalam penggunaan jaringan saraf tiruan (Neurak Network) yang sangat dalam (dalamhal jumlah lapisan/layer yang dalam). Tujuannya untuk mempelajari representasi data yang semakin abstrak dan kompleks.

# Metode Deep Learning

## Feedforward Neural Network

Merupakan Jaringan Saraf Tiruan (Neural Network) yang terdiri dari lapisan - lapisan. Tujuannya adalah untuk tugas seperti klasifikasi, regresi dan pemrosesan data(data processing)

## Convolutional Neural Network (CNN)

CNN biasanya digunakan untuk tugas - tugas pengelolaan citra dan visual. Jadi metode ini menggunakan lapisan konvolusi untuk mengidentifikasi fiture-fiture spesial dalam citra.

## Generative Adversarial Network (GAN)

GAN memiliki dua jaringan, yaitu generator dan discriminator yang bersaing satu sama lain. Hal ini digunakan untuk menghasilkan data baru yang mirip dengan data yang ada, seperti generasi gambar.

## Autoencoders

Merupakan arsitektur yang digunakan untuk pembelajaran tanpa pengawasan dan reduksi dimensi data (Principal Component Analysis).

## Matplotlib

Tujuan : untuk membuat visualisasi statistik, interaktif, dan animasi.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Grafik Sin')
plt.show()
```
